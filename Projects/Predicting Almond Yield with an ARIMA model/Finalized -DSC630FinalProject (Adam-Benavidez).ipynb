## Train and Test Datasets

```python
# Train-Test Split for SetOne (1961 - 2021, 60 years)
train_SetOne = SetOne[SetOne['Year'] < 2011] # 50 years
test_SetOne = SetOne[SetOne['Year'] >= 2011] # 10 years

# Train-Test Split for SetTwo (1991 - 2021, 30 years)
train_SetTwo = SetTwo.loc[SetTwo['Year'] < 2016] # 25 years 
test_SetTwo = SetTwo.loc[SetTwo['Year'] >= 2016] # 5 years
```

## Time Series Forecasting Models

```python
# Load Necessary libraries for models 
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error
import statsmodels.api as sm
import itertools
from statsmodels.tools.sm_exceptions import ConvergenceWarning

# Disable the specific warning
warnings.filterwarnings("ignore", message="Maximum Likelihood optimization failed to converge. Check mle_retvals", category=ConvergenceWarning)
warnings.filterwarnings("ignore", message="An unsupported index was provided and will be ignored when e.g. forecasting.", category=UserWarning)
```

```python
# Determine best ARIMA Order for SetOne
best_aic_SetOne = np.inf
best_order_SetOne = None

for p in range(3):
    for d in range(2):
        for q in range(3):
            try:
                model = sm.tsa.ARIMA(SetOne['Year'], order=(p, d, q))
                results = model.fit()
                if results.aic < best_aic_SetOne:
                    best_aic_SetOne = results.aic
                    best_order_SetOne = (p, d, q)
            except:
                continue

print('Best ARIMA order for SetOne:', best_order_SetOne)
print('Best AIC for SetOne:', best_aic_SetOne)
```

```python
# Determine best ARIMA Order for SetTwo
best_aic_SetTwo = np.inf
best_order_SetTwo = None

for p in range(3):
    for d in range(2):
        for q in range(3):
            try:
                model = sm.tsa.ARIMA(SetTwo['Year'], order=(p, d, q))
                results = model.fit()
                if results.aic < best_aic_SetTwo:
                    best_aic_SetTwo = results.aic
                    best_order_SetTwo = (p, d, q)
            except:
                continue

print("Best ARIMA order for SetTwo:", best_order_SetTwo)
print("Best AIC for SetTwo:", best_aic_SetTwo)
```

```python
# ARIMA models for SetOne
model_SetOne = {}
results_SetOne = {}

for col in train_SetOne.columns:
    model_SetOne = sm.tsa.ARIMA(train_SetOne[col], order=(2, 1, 1))
    results_SetOne[col] = model_SetOne.fit()

# ARIMA models for SetTwo
model_SetTwo = {}
results_SetTwo = {}

for col in train_SetTwo.columns:
    model_SetTwo = sm.tsa.ARIMA(train_SetTwo[col], order=(2, 1, 2))
    results_SetTwo[col] = model_SetTwo.fit()

# Print the results for SetOne
for col, result in results_SetOne.items():
    print(f"Results for SetOne {col}: ")
    print(result.summary())
    print("---------------------------")

# Print the results for SetTwo
for col, result in results_SetTwo.items():
    print(f"Results for SetTwo{col}: ")
    print(result.summary())
    print("---------------------------")
```

```python
# Forecast and Evaluate
forecast_SetOne = {}
forecast_SetTwo = {}

# Forecast for SetOne
for col, model in results_SetOne.items():
    forecast_SetOne[col] = model.forecast(steps=11)

# Forecast for SetTwo
for col, model in results_SetTwo.items():
    forecast_SetTwo[col] = model.forecast(steps=6)
```

## Compare Performance Metrics

```python
#drop 'Year' column
train_SetOne.drop(['Year'], axis=1, inplace=True)
test_SetOne.drop(['Year'], axis=1, inplace=True)

train_SetTwo.drop(['Year'], axis=1, inplace=True)
test_SetTwo.drop(['Year'], axis=1, inplace=True)
```

```python
# Compare Performance
mae_SetOne = {}
mse_SetOne = {}
rmse_SetOne = {}
mae_SetTwo = {}
mse_SetTwo = {}
rmse_SetTwo = {}

# Calculate metrics for SetOne
for col in test_SetOne.columns:
    mae_SetOne[col] = mean_absolute_error(test_SetOne[col], forecast_SetOne[col])
    mse_SetOne[col] = mean_squared_error(test_SetOne[col], forecast_SetOne[col])
    rmse_SetOne[col] = np.sqrt(mse_SetOne[col])
    
# Calculate metrics for SetTwo
for col in test_SetTwo.columns:
    mae_SetTwo[col] = mean_absolute_error(test_SetTwo[col], forecast_SetTwo[col])
    mse_SetTwo[col] = mean_squared_error(test_SetTwo[col], forecast_SetTwo[col])
    rmse_SetTwo[col] = np.sqrt(mse_SetTwo[col])    
    
# Print the performance metrics for SetOne
print("Performance metrics for SetOne:")
for col in test_SetOne.columns:
    print(f"Column '{col}':")
    print(f"MAE: {mae_SetOne[col]}")
    print(f"MSE: {mse_SetOne[col]}")
    print(f"RMSE: {rmse_SetOne[col]}\n")

# Print the performance metrics for SetTwo
print("Performance metrics for SetTwo:")
for col in test_SetTwo.columns:
    print(f"Column '{col}':")
    print(f"MAE: {mae_SetTwo[col]}")
    print(f"MSE: {mse_SetTwo[col]}")
    print(f"RMSE: {rmse_SetTwo[col]}\n")
```
